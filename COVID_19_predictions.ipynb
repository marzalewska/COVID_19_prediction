## Analiza globalna

Jako poczatkową analizę stworzony zostanie ranking państw, w których nastąpiło najwięcej i najmniej zgonów przypadających na milion mieszkańców. Następnie stworzone zostały wykresy ilustrujące jak szybko rozwijała się pandemia w tych krajach. Do wykresów potrzebne były pewne modyfikacje danych prawdopodobnie ze względu na błędy informacji podawanych do wiadomości publicznych bądź błędy systemu pobierającego dane.

```python
grouped_by_country = covid.groupby('administrative_area_level_1')
death_counts = grouped_by_country['deaths'].max()
populations = grouped_by_country['population'].max()
deaths_per_million = (death_counts / populations)*1000000
deaths_per_million = deaths_per_million[deaths_per_million != 0.0]
pd.options.display.float_format = '{:.1f}'.format

display(deaths_per_million.sort_values().head(5))
display(display(deaths_per_million.dropna().sort_values().tail(5)))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

countries_min_deaths = deaths_per_million.sort_values().head(5).index
plt.figure(figsize=(14, 8))
for country in countries_min_deaths:
  country_data=covid[covid['administrative_area_level_1']==country]
  plt.plot(country_data['date'], country_data['deaths'], label=country, lw=3)
plt.xlabel('Data')
plt.ylabel('Zgony')
plt.title('Kraje z najmniejszą liczbą zgonów na milion ludzi z powodu COVID-19')
plt.legend()
plt.grid(True)
plt.show()
```

```python
countries_max_deaths = deaths_per_million.dropna().sort_values().tail(5).index
plt.figure(figsize=(14, 8))
for country in countries_max_deaths:
  country_data=covid[covid['administrative_area_level_1']==country]
  if country=='Peru':
    country_data[country_data['date']=='31.05.2021']=country_data[country_data['date']=='30.05.2021']
  if country=='Croatia':
    country_data[country_data['date']=='14.10.2022']=country_data[country_data['date']=='13.10.2022']
  plt.plot(country_data['date'], country_data['deaths'], label=country, lw=3)
plt.xlabel('Data')
plt.ylabel('Zgony')
plt.title('Kraje z największą liczbą zgonów na milion ludzi z powodu COVID-19')
plt.legend()
plt.grid(True)
plt.show()
```

## Regresja liniowa

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

columns_to_death_regr = brazylia.columns[9:23]
death_collective_regr = []

for var in columns_to_death_regr:
    X = brazylia[[var]]
    y = brazylia['deaths']

    model = LinearRegression()
    model.fit(X, y)

    y_pred = model.predict(X)
    mse = mean_squared_error(y, y_pred)
    r2 = r2_score(y, y_pred)

    death_collective_regr.append({
        'Zmienna': var,
        'Współczynnik': model.coef_[0],
        'Odcięta': model.intercept_,
        'MSE': mse,
        'R^2': r2
    })

death_collective_regr = pd.DataFrame(death_collective_regr)
death_collective_regr = death_collective_regr.sort_values(by=['R^2', 'MSE'], ascending=[False, True])
display(death_collective_regr)
```

```python
columns_to_confirmed_regr=brazylia.columns[5:23]
confirmed_collective_regr = []

for var in columns_to_confirmed_regr:
    X = brazylia[[var]]
    y = brazylia['confirmed']

    model = LinearRegression()
    model.fit(X, y)

    y_pred = model.predict(X)
    mse = mean_squared_error(y, y_pred)
    r2 = r2_score(y, y_pred)

    confirmed_collective_regr.append({
        'Zmienna': var,
        'Współczynnik': model.coef_[0],
        'Odcięta': model.intercept_,
        'MSE': mse,
        'R^2': r2
    })

confirmed_collective_regr = pd.DataFrame(confirmed_collective_regr)
confirmed_collective_regr = confirmed_collective_regr.sort_values(by=['R^2', 'MSE'], ascending=[False, True])
display(confirmed_collective_regr)
```

## regresja wielomianowa

Dla dodatkowego sprawdzenia korelacji dodane zostały zmienne liczące dzienny przyrost przypadków zakażenia, śmierci, osób zaszczepionych oraz zrobionych testów.

```python
brazylia['daily_conf_increase'] = brazylia['confirmed'].diff()
brazylia['daily_deaths_increase'] = brazylia['deaths'].diff()
brazylia['daily_vaccines'] = brazylia['vaccines'].diff()
brazylia['daily_tests'] = brazylia['tests'].diff()
brazylia=brazylia.fillna(0)

columns_to_correlate = brazylia.columns[2:23].tolist()+['daily_conf_increase', 'daily_deaths_increase','daily_vaccines','daily_tests']
correlation_pearson = brazylia[columns_to_correlate].corr(method='pearson')
correlation_spearman = brazylia[columns_to_correlate].corr(method='spearman')

fig, ax = plt.subplots(1, 2, figsize=(24, 10))

sns.heatmap(correlation_pearson, annot=True, cmap='coolwarm', vmin=-1, vmax=1, cbar=True, ax=ax[0],annot_kws={"size": 8},fmt=".2f")
ax[0].set_title('Macierz korelacji Pearsona')

sns.heatmap(correlation_spearman, annot=True, cmap='coolwarm', vmin=-1, vmax=1, cbar=True, ax=ax[1],annot_kws={"size": 8},fmt=".2f")
ax[1].set_title('Macierz korelacji Spearmana')

plt.show()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

X = brazylia[columns_to_correlate]
X = X.assign(const=1)

vif_data = pd.DataFrame()
vif_data["Variable"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
print(vif_data)
```

## Algorytmy SVR, drzew regresyjnych i losowego lasu regresyjnego

```python
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler


col_regression_models = ['recovered', 'tests', 'vaccination_policy', 'people_vaccinated']

X = brazylia[col_regression_models]
y = brazylia.loc[X.index, 'deaths']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=65)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

svr_model = SVR()
tree_model = DecisionTreeRegressor()
forest_model = RandomForestRegressor()

svr_model.fit(X_train_scaled, y_train)
tree_model.fit(X_train, y_train)
forest_model.fit(X_train, y_train)

svr_predictions = svr_model.predict(X_test_scaled)
tree_predictions = tree_model.predict(X_test)
forest_predictions = forest_model.predict(X_test)

svr_mse = mean_squared_error(y_test, svr_predictions)
tree_mse = mean_squared_error(y_test, tree_predictions)
forest_mse = mean_squared_error(y_test, forest_predictions)

svr_r2 = r2_score(y_test, svr_predictions)
tree_r2 = r2_score(y_test, tree_predictions)
forest_r2 = r2_score(y_test, forest_predictions)

print("SVR:")
print(f"MSE: {svr_mse}")
print(f"R^2: {svr_r2}\n")

print("Drzewo regresyjne:")
print(f"MSE: {tree_mse}")
print(f"R^2: {tree_r2}\n")

print("Losowy las regresyjny:")
print(f"MSE: {forest_mse}")
print(f"R^2: {forest_r2}")
```
